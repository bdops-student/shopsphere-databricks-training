```python
import pyspark.sql.functions as F
orders = spark.table("shopsphere_catalog.retail.orders")
print(f"Orders total: {orders.count():,}")
print(f"Date range: {orders.agg(F.min('order_purchase_timestamp'), F.max('order_purchase_timestamp')).collect()}")
orders.groupBy("order_status").count().orderBy("count", ascending=False).show()
```
